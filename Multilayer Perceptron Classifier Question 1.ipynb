<b> Question-1: </b><br>

Implement a perceptron model to realize AND gate. Start with the weights
as [0.5, 0.5]. <br> Update the weights using perceptron learning algorithm.
Print the model parameters after every epoch. Use suitable learning rate.

<b>Model Implementation: </b>

```python
# Importing necessary modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
def perceptron(inputs, weights, learning_rate):
  # Include bias term as the first element in the weight vector
  bias = weights[0]

  # Calculate weighted sum using the updated weights
  z = np.dot(inputs, weights[1:]) + bias

  # Apply the activation function (step function)
  predictions = np.where(z >= 0, 1, 0)

  return predictions

def update_weights(inputs, targets, predictions, weights, learning_rate):
  # Calculate the error for each sample
  errors = targets - predictions

  # Update weights based on the learning rule
  for i in range(len(inputs)):
    for j in range(len(weights)):
      # Update weight excluding the bias term
      if j > 0:
        weights[j] += learning_rate * errors[i] * inputs[i][j - 1]
      else:
        # Update bias separately (no input dependency)
        weights[j] += learning_rate * errors[i]
```

<b> Application of Model to the Test Case: </b>

```python
# Define the AND gate inputs and targets
inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
targets = np.array([0, 0, 0, 1])

# Initialize weights, including bias as the first element bias = 1 and weights initialized as 0.5 each
weights = np.array([1, 0.5, 0.5])

# Set learning rate
learning_rate = 0.05

epochs = 0
flag = 0

# Train the perceptron model
while(flag != 1):
  # Update the number of epochs
  epochs+=1

  # Make predictions
  predictions = perceptron(inputs, weights, learning_rate)

  # Print model parameters
  print(f"Epoch {epochs}:")
  print(f"  Bias: {weights[0]}")
  print(f"  Weights: {weights[1:]}")
  print(f"  Predictions: {predictions}")
  print(f"  Targets: {targets}")
  print("")

  # Checking if targets are met or not
  f = 0
  for i in range(4):
    if(targets[i] != predictions[i]):
      f = 1; break
  if(f == 1): flag = 0
  else: flag = 1

  # Update weights
  update_weights(inputs, targets, predictions, weights, learning_rate)

print(f"Hence, after {epochs} epochs we have arrived at our target value through the perceptron model.")
```
