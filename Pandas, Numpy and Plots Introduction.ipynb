<b> 8) Visualize distribution of 3 different classes in the iris dataset.

```python
#visual distribution by count plot for the dataset

sns.countplot(x="Species",data=dt)
plt.title("Visual distribution of 3 different classes in the Iris dataset\n")
plt.show()
```

<b> 9) Plot the relationship between three different iris classes with respect to
petal length and petal width.

```python
#making scatterplot for displaying relationship required
sns.scatterplot(x="PetalLengthCm",y="PetalWidthCm",hue="Species",data=dt)
plt.title("Plot of relation b/w petal length and petal width in the Iris dataset\n")
plt.show()
```
