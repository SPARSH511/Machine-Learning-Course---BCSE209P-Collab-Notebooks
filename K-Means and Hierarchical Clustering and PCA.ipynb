# **shop.csv Dataset Analysis using Cluster Models**

```python
# Read the dataset shop.csv

df3 = pd.DataFrame(pd.read_csv("shop.csv"))
```

```python
# Checking for the kind of values in dataset

df3.head(10)
```

```python
# Checking for duplicates

df3.duplicated().sum()
```

```python
# Checking for null values

df3.info()
```

```python
# Dropping the CustomerID (not relevant to clustering part)

df3.drop("CustomerID",axis=1,inplace=True)
df3.head(10)
```

```python
# Label Encoding for the categorical columns

le = LabelEncoder()
df3['Gender'] = le.fit_transform(df3['Gender'])

df3.head(10)
```

```python
# Separating the 'cluster' column from rest of the data

X2 = df3.drop('cluster',axis=1)
y2 = df3['cluster']
```

```python
# Elbow Method Implementation to know the optimum number of clusters

cluster_numbers = range(2,10)
inertia = []
silhouette_scores = []
rand_scores = []

for k in cluster_numbers:
    kmeans_model = KMeans(n_clusters=k,random_state=42).fit(X2)
    inertia.append(kmeans_model.inertia_)
    silhouette_avg = silhouette_score(X2, kmeans_model.labels_)
    silhouette_scores.append(silhouette_avg)
```

```python
# We will observe the chart to define an elbow

plt.plot(cluster_numbers, inertia, marker='o')
plt.title("Intertia Score Chart\n")
plt.xlabel("No. of clusters")
plt.ylabel("Inertia Score")
plt.show()
```

```python
# Silhouette Score Chart to help us find the elbow

plt.plot(cluster_numbers, silhouette_scores, marker='o')
plt.title("Silhouette Score Chart\n")
plt.xlabel("No. of clusters")
plt.ylabel("Silhouette Score")
plt.show()
```

**After the observation of all the charts we can conclusively say that elbow lies at 6 as the no. of clusters because of heuristics of the minimal change point in inertia and maximal Silhouette score. So, we have found the value of n_clusters = 6.**

```python
# Model Formation for KMeans Clustering

kmeans_model = KMeans(n_clusters=6, random_state=42)
kmeans_model.fit(X2)
predictions = kmeans_model.fit_predict(X2)
```

```python
# Model Evaluation for KMeans Clustering

print("Rand Score: ",rand_score(y2,predictions))
print("Silhouette Score: ",silhouette_score(X2,predictions))
print("Adjusted Mutual Information Score: ",adjusted_mutual_info_score(y2,predictions))
print("V-Measure Score: ",v_measure_score(y2,predictions))
print("Fowlkes Mallows Score: ",fowlkes_mallows_score(y2,predictions))
```

```python
# Model Formation for Hierarchical Clustering

hier_model = AgglomerativeClustering(n_clusters=6)
hier_model.fit(X2)
predictions1 = hier_model.fit_predict(X2)
```

```python
# Model Evaluation for Hierarchical Clustering

print("Rand Score: ",rand_score(y2,predictions1))
print("Silhouette Score: ",silhouette_score(X2,predictions1))
print("Adjusted Mutual Information Score: ",adjusted_mutual_info_score(y2,predictions1))
print("V-Measure Score: ",v_measure_score(y2,predictions1))
print("Fowlkes Mallows Score: ",fowlkes_mallows_score(y2,predictions1))
```

**Hence, we can observe that KMeans Clustering provides us with 1.0 scores in most of the metrics which means the classification is perfect as the scale of these scores is from 0 to 1 or -1 to 1 (in case of Adjusted Mutual Information Score) where 1 represents perfect classification into clusters and 0 represents as bad it can be and negative values represent totally wrong clusters.**

**Hierarchical Clustering does not perform that well in comparision but still has a decent performance on the dataset.**

# **Dataset Analysis after PCA Application on the dataset shop.csv**

```python
# We try to apply PCA to reduce the number of columns to 2 for better visualization

pca = PCA(n_components = 2)
X2_pca = pd.DataFrame(pca.fit_transform(X2))

X2_pca.head(10)
```

```python
# Cluster Scatterplot for the given data - True Labels

curr_data = X2_pca.assign(cluster=y2)

sns.scatterplot(x=0, y=1, hue='cluster', data=curr_data, palette=["red","green","blue","yellow","orange","magenta"])
plt.xlabel("PC0",fontweight="bold")
plt.ylabel("PC1",fontweight="bold")
plt.title("Scatterplot with the provided true labels\n",fontweight="bold")
plt.legend(title='cluster')
plt.show()
```

```python
# Model Formation for KMeans Clustering

kmeans_model = KMeans(n_clusters=6, random_state=42)
kmeans_model.fit(X2_pca)
predictions = kmeans_model.fit_predict(X2_pca)
```

```python
# Model Evaluation for KMeans Clustering

print("Rand Score: ",rand_score(y2,predictions))
print("Silhouette Score: ",silhouette_score(X2_pca,predictions))
print("Adjusted Mutual Information Score: ",adjusted_mutual_info_score(y2,predictions))
print("V-Measure Score: ",v_measure_score(y2,predictions))
print("Fowlkes Mallows Score: ",fowlkes_mallows_score(y2,predictions))
```

```python
# Cluster Scatterplot for the given data - Predicted Labels - KMeans

curr_data = X2_pca.assign(cluster=predictions)

sns.scatterplot(x=0, y=1, hue='cluster', data=curr_data, palette=["red","green","blue","yellow","orange","magenta"])
plt.xlabel("PC0",fontweight="bold")
plt.ylabel("PC1",fontweight="bold")
plt.title("Scatterplot with the provided predicted labels - KMeans\n",fontweight="bold")
plt.legend(title='cluster')
plt.show()
```

```python
# Model Formation for Hierarchical Clustering

hier_model = AgglomerativeClustering(n_clusters=6)
hier_model.fit(X2_pca)
predictions1 = hier_model.fit_predict(X2_pca)
```

```python
# Model Evaluation for Hierarchical Clustering

print("Rand Score: ",rand_score(y2,predictions1))
print("Silhouette Score: ",silhouette_score(X2_pca,predictions1))
print("Adjusted Mutual Information Score: ",adjusted_mutual_info_score(y2,predictions1))
print("V-Measure Score: ",v_measure_score(y2,predictions1))
print("Fowlkes Mallows Score: ",fowlkes_mallows_score(y2,predictions1))
```
